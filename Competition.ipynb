```python
from pylab import *
from scipy.integrate import odeint

%matplotlib inline
```

```python
def model(x, t):
    x0dot = x[0] * (1 - x[0] - a * x[1])
    x1dot = x[1] * (1 - a * x[0] - x[1])
    return [x0dot, x1dot]
```

```python
a = .9
t = linspace(0, 100, 10000)
```

```python
x = odeint(model, [0, 0.01], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.01, .01], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.02, .01], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.05, .01], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.01, .01], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.01, .02], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.01, .05], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [0.01, .0], t)
plot(x[:,0], x[:,1], color = 'red')

x = odeint(model, [2., 0], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [2., 0.2], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [2., 0.5], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [2., 1.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [2., 1.5], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [2., 2.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [1.5, 2.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [1., 2.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [.5, 2.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [.2, 2.], t)
plot(x[:,0], x[:,1], color = 'green')

x = odeint(model, [0, 2.], t)
plot(x[:,0], x[:,1], color = 'green')

axis([-.1, 2.1, -.1, 2.1])
```
