```python
from pylab import *
from scipy.stats import binom
%matplotlib inline
```

```python
N = 10000
p =.05
n = arange(N + 1)
pmf = binom.pmf(n, N, p)
vlines(n, 0, pmf)
#axis([-.1,100.1,0,.5])
```

```python
N = 2
p =.5
n = arange(N + 1)
sampSize = 2000
pmf = binom.pmf(n, N, p)
sample = binom.rvs(N, p, size = sampSize)
hist, binEdges = histogram(sample, bins = arange(N +2)-.5, density = True)
vlines(n, 0, hist)
plot(n, pmf)
axis([-.1, 2.1, 0, 1])
```
