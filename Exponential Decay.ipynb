```python
from pylab import *
from scipy.stats import expon
%matplotlib inline
```

```python
def simula(N0):
    t = zeros(1)
    N = array([N0])
    while N[-1] > 0:
        a = gamma * N[-1]
        tau = expon.rvs(scale = 1. / a)
        t = append(t, [t[-1] + tau, t[-1] + tau])
        N = append(N, [N[-1], N[-1] - 1])
    return t, N
```

```python
gamma = 1.
N0 = 10
tmax = 0

t, N = simula(N0)
tmax = max(tmax, t[-1])
plot(t, N)

t, N = simula(N0)
tmax = max(tmax, t[-1])
plot(t, N)

t, N = simula(N0)
tmax = max(tmax, t[-1])
plot(t, N)

tt = linspace(0, tmax, 200)
plot(tt, N0*exp(-gamma * tt), color = 'black')
```
